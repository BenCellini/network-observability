# Custom colors set manually

```python
# Create network visualizer & draw
layer_sizes  = (8, 6, 4, 2, 4, 6, 8)
network = FCNNVisualizer(layers=layer_sizes,
                         neuron_radius=0.3,
                         layer_space=2.0,
                         connection_alpha=0.5,
                         connection_linewidth=1.0,
                         neuron_edgewidth=2)
# Set some neuron colors
network.neuron_colors[0] = 'r'  # set the first neuron color
network.neuron_colors[int(network.n_neuron/2)] = 'g'  # set the midway neuron color
network.neuron_colors[network.n_neuron-1] = 'b'  # set the last neuron color

# Set some neuron edge colors
network.neuron_edge_colors[3] = 'c'
network.neuron_edge_colors[15] = 'm'
network.neuron_edge_colors[28] = 'y'

# Set some connection colors
for n in range(0, layer_sizes[0]*layer_sizes[1]):  # for the entire first layer
    network.connection_colors[n] = 'purple'
    
for n in range(layer_sizes[0]*layer_sizes[1], layer_sizes[0]*layer_sizes[1] + layer_sizes[1]*layer_sizes[2]):  # for the next layer
    network.connection_colors[n] = 'orange'

# Draw
network.draw()
```

# Custom colors set based on values

```python
# Create network visualizer & draw
layer_sizes  = (8, 6, 4, 2, 4, 6, 8)
network = FCNNVisualizer(layers=layer_sizes,
                         neuron_radius=0.3,
                         layer_space=2.0,
                         connection_alpha=0.5,
                         connection_linewidth=1.0,
                         neuron_edgewidth=2)
# Set colors
neuron_colors = np.linspace(0, 1, num=network.n_neuron)
connection_colors = np.linspace(1, 2, num=network.n_connection)

network.set_colors(name='neuron', values=neuron_colors, colormap='inferno', norm=None)  # scale neuron color by value and colormap
network.set_colors(name='connection', values=connection_colors, colormap='plasma_r', # set a log normalized colormap for neuron connections
                   norm=mpl.colors.LogNorm(vmin=np.min(connection_colors), vmax=np.max(connection_colors)))
network.set_colors(name='neuron_edge', colors='gray')  # make all neuron edges one color

# Draw
network.draw()
```

# Set all centers to ellipsis

```python
# Create network visualizer & draw
layer_sizes  = (7, 5, 3, 5, 7)
network = FCNNVisualizer(layers=layer_sizes,
                         neuron_radius=0.3,
                         layer_space=2.0,
                         connection_alpha=0.5,
                         connection_linewidth=1.0,
                         neuron_edgewidth=2)

network.set_layer_continuation(neuron_index='center', no_ends=True) # no_ends set to True excludes the input and output layers

# Draw
network.draw()
```

# Set the first layer to have a pattern of reacting ellipsis

```python
# Create network visualizer & draw
layer_sizes  = (12, 15, 17, 15, 1)
network = FCNNVisualizer(layers=layer_sizes,
                         neuron_radius=0.3,
                         layer_space=2.0,
                         connection_alpha=0.5,
                         connection_linewidth=1.0,
                         neuron_edgewidth=1)

network.set_layer_continuation(neuron_index='center', no_ends=True)
network.set_layer_continuation(neuron_index='pattern', params={'start': 1, 'on': 2, 'off': 2, 'layers': (0,)}, no_ends=False)  # params values specify the spacing pattern

# Draw
network.draw()
```
